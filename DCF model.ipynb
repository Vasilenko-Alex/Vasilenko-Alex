```python

import pandas as pd

years = [2020, 2021, 2022]

#Change in WC Calculation 

def change_in_wc():

    inventory = [150 , 300, 350, 500]
    accounts_payable = [200, 300, 500, 400 ]
    accounts_receivable = [100, 250, 300, 400 ]
    years_wc = [2019,2020,2021,2022]
    change_in_nwc_p_y = []
    previous_y_nwc = []

    for i in range(len(years_wc)):
        calculation = inventory[i] + accounts_receivable[i] - accounts_payable[i]
        previous_y_nwc.append(calculation)
        if i > 0: 
            change_in_nwc_p_y.append(calculation - previous_y_nwc[i -1])
    return change_in_nwc_p_y

WC = change_in_wc()

#Free Cash Flow Calculation

def free_cash_flow_to_firm(WC):

    net_income = [1000, 1000, 1000]
    depr = [200, 300, 400]
    int_exp = [300,400, 450]
    capex = [500 , 600, 700 ]
    tax_rate = 0.30

    FCFF_per_year = []
    for ni, de, inter, wc , cpx in zip(net_income, depr, int_exp, WC, capex):
        calculation = ni + de + (inter * (1 - tax_rate)) - wc - cpx
        FCFF_per_year.append(calculation)
    return FCFF_per_year

FCFF = free_cash_flow_to_firm(WC)

# Cost of Capital Calculation 

def capm():
    beta = 1.1
    risk_free_rate = 0.04
    expected_market_return = 0.08
    CAPM = risk_free_rate + beta * (expected_market_return - risk_free_rate) 
    return CAPM
cost_of_equity = capm()

def wacc(cost_of_equity):
    debt_to_total_capital = 0.7
    cost_of_debt = 0.05 
    tax_rate = 0.3
    calculation = ((cost_of_debt * (1 - tax_rate)) * debt_to_total_capital) + ((1 - debt_to_total_capital) * cost_of_equity)
    return calculation

WACC = wacc(cost_of_equity) 

def disc_fact(WACC):
    disc_f_y = []
    for i in range(len(years)):
        calc = (1 + WACC ) ** (i + 1)
        disc_f_y.append(calc)
    return disc_f_y

disc_factor = disc_fact(WACC)

#Discounting the FCFF

def DCF(FCFF):
    terminal_multiple = 3
    DCF_p_y = []
    terminal_value = FCFF [- 1 ] * terminal_multiple  / disc_factor [- 1] 
    for i in range(len(years)):
        calculation = round (FCFF[i] / disc_factor[i], 2)
        DCF_p_y.append(calculation)
        fully_disc_value = round(sum(DCF_p_y) + terminal_value,2)
    return DCF_p_y, terminal_value , fully_disc_value

result, terminal_val,fully_disc_value = DCF(FCFF)

def fair_value_per_share(fully_disc_value):
    market_value_of_debt = 1000
    shares_outst = 100
    value_of_equity = round(fully_disc_value - market_value_of_debt, 2)
    fair_value_per_share = round(value_of_equity / shares_outst,2)
    return fair_value_per_share, value_of_equity

FV, value_of_equity = fair_value_per_share(fully_disc_value)

#Pandas table
     
table = pd.DataFrame({"Free_Cash_Flow_to_Firm": FCFF, "Discounted_CF_per_Year": result}, index = years)
sum_row = table.sum()
table_with_sum = table.append(sum_row, ignore_index= True)
table_with_sum.index = years + ['Sum']
table_with_sum.index.name = "Years"
print (table_with_sum)
print()
print(f"Fully_Discounted_value $mln: {fully_disc_value}")
print(f"Value_of_Equity $mln: {value_of_equity}") 
print(f"Value $ per / Share: {FV}")



```
