```python
#Working

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

Years = [2018, 2019, 2020, 2021, 2022, 2023, 2024]
EBITDA = [None, None, None, None, 130541.0, 127106.77, 137501.81]
EV_EBITDA_5Y = [12.89, 12.25, 24.53, 19.38, 17.09]
avg_EV_EBITDA = np.mean(EV_EBITDA_5Y)

def EV_EBITDA(EBITDA, Years):
    market_cap_of_equity = 2838000.13
    ST_debt = 7216.0 + 3993
    LT_debt = 98071.0
    cash_and_ST_investments = 62482.0
    EV_EBITDA_multiple = []

    for year in range(len(Years)):
        if EBITDA[year] is not None:
            calculation = round(
                (market_cap_of_equity + ST_debt + LT_debt - cash_and_ST_investments) / EBITDA[year], 2)
            EV_EBITDA_multiple.append(calculation)
        else:
            EV_EBITDA_multiple.append(None)
    return EV_EBITDA_multiple

valuation = EV_EBITDA(EBITDA, Years)
trailing_12M = valuation[-3]
one_y_fwd = valuation[-2]
two_y_fwd = valuation[-1]
historic_years = Years[:len(EV_EBITDA_5Y)]

table = pd.DataFrame({
        "Years": historic_years + ["1Y_Forward"] + ["2Y_Forward"],
        "EV/EBITDA multiple X": EV_EBITDA_5Y  + [one_y_fwd] + [two_y_fwd],})
        
table.set_index("Years", inplace=True)
print(table)
print()
print(f'5_Y_avg_multiple X:  {avg_EV_EBITDA}')


plt.plot(Years, table["EV/EBITDA multiple X"], marker="o", label = "Multiple_X")
plt.xlabel("Years")
plt.ylabel("EV/EBITDA")
plt.title("EV/EBITDA Valuation Over Time")
plt.legend()
plt.xticks(Years)
plt.grid(True)
plt.tight_layout()
plt.show()




```
